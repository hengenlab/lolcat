```python
import pandas as pd
from os import listdir
import numpy as np
```

```python
#hp_grid_files = [x for x in listdir('.') if 'hp_grid' in x and '.csv' in x]
hp_grid_files = ['hp_grid_p5.csv']
hp_grid_df = pd.concat([pd.read_csv(x) for x in hp_grid_files]).sort_values(by='hp_idx')
hp_grid_idxs = hp_grid_df.hp_idx.tolist()
hp_grid_f1_files = ['ISI_hp{}_f1s_17celltypes.csv'.format(str(x)) for x in hp_grid_idxs if 'ISI_hp{}_f1s_17celltypes.csv'.format(str(x)) in listdir('.')]
found_hp_grid_idxs = [x for x in hp_grid_idxs if 'ISI_hp{}_f1s_17celltypes.csv'.format(str(x)) in listdir('.')]
hp_grid_df = hp_grid_df[hp_grid_df['hp_idx'].isin(found_hp_grid_idxs)]
hp_grid_f1s_df = [pd.read_csv(x) for x in hp_grid_f1_files]
print(hp_grid_f1s_df)
hp_grid_train_max_f1s = [np.max(df.train_acc.to_numpy()) for df in hp_grid_f1s_df]
hp_grid_test_max_f1s = [np.max(df.test_acc.to_numpy()) for df in hp_grid_f1s_df]
hp_grid_val_max_f1s = [np.max(df.val_acc.to_numpy()) for df in hp_grid_f1s_df]
hp_grid_df['train_f1'] = hp_grid_df['hp_idx'].map(dict(zip(hp_grid_idxs,hp_grid_train_max_f1s)))
hp_grid_df['test_f1'] = hp_grid_df['hp_idx'].map(dict(zip(hp_grid_idxs,hp_grid_test_max_f1s)))
hp_grid_df['val_f1'] = hp_grid_df['hp_idx'].map(dict(zip(hp_grid_idxs,hp_grid_val_max_f1s)))
hp_grid_df.to_csv('compiled_hp_grid_regularization.csv',index=False)
```
